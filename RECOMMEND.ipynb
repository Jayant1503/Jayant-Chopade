```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation 
```

```python
book=pd.read_csv("book (1).csv",encoding = 'ISO-8859-1')
book
```

```python
books=book.iloc[:,1:]
books
```

```python
# Sorting by User IDs
books.sort_values(['User.ID'])
```

```python
# Renaming the column
books = books.rename({'User.ID':'UserId','Book.Title':'Title','Book.Rating':'Rating'}, axis=1)
books
```

```python
# Number of UserId ,Title and Ratings
books.nunique()
```

```python
# Number of Ratings ad counts
books['Rating'].value_counts().sort_index()
```

```python
# Plotting the rating in the barplot
books['Rating'].value_counts().sort_index().plot(kind='bar')
```

```python
plt.figure(figsize=(18,6))
sns.distplot(books.Rating)
```

```python
# using PIVOT table TO CONVERT LONG DATA INTO WIDE DATA
books2=books.pivot_table(index='UserId',columns='Title',values='Rating').reset_index(drop=True)
books2
```

```python
#replacing with unique user id
books2.index=books['UserId'].unique()
books2
```

```python
# filling missing values with 0 values
books2.fillna(0,inplace=True)
books2
```

```python
# Calculating Cosine Similarity between Users on array data
usersim=1-pairwise_distances(books2.values,metric='cosine')
usersim
```

```python
# Storing the results in a dataframe format
usersim2=pd.DataFrame(usersim)
usersim2
```

```python
# Set the index and column names to UserId
usersim2.index=books['UserId'].unique()
usersim2.columns=books['UserId'].unique()
usersim2
```

```python
# Nullifying diagonal values
np.fill_diagonal(usersim,0)
usersim2
```

```python
# Most Similar Users
usersim2.idxmax(axis=1)
```

```python
def get_recommendation(customer_id):
        
        t=list(usersim2.sort_values([customer_id],ascending=False).head(3).index)

        book_list=list(books[books['UserId'].isin(t)]['Title'])

        recommended_books = set(book_list)-set(books[books['UserId']==customer_id]['Title'])
        
        return recommended_books
```

```python
#getting recommendations
get_recommendation(162113)
```
