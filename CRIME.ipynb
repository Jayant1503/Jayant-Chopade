```python
#importing libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime=pd.read_csv('crime_data.csv')
crime
```

```python
#Normalized data fuction
def norm_function(i):
    f=(i-i.min())/(i.max()-i.min())
    return(f)
```

```python
datanorm=norm_function(crime.iloc[:,1:])
datanorm
```

```python
dendrogram=sch.dendrogram(sch.linkage(datanorm,method='average'))
```

```python
# creating clusters
cc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
cc
```

```python
y=cc.fit_predict(datanorm)
```

```python
y
```

```python
crime['h_clusterid']=cc.labels_
crime
```

```python
#k means clustering
crimedata=pd.read_csv('crime_data.csv')
```

```python
#Normalized data fuction
def nofunc(i):
    m=(i-i.min())/(i.max()-i.min())
    return(m)
```

```python
datanorm=norm_function(crime.iloc[:,1:])
datanorm
```

```python
# Elbow curv
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(datanorm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(datanorm)
model.labels_
```

```python
x=pd.Series(model.labels_)
crimedata['Clust']=x
```
